```python
# EDA_preBI.ipynb
# Exploratory Data Analysis sobre las tablas finales del pipeline

import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

plt.style.use("ggplot")

CWD = os.getcwd()
DATA_FINAL = os.path.join(CWD, "data_final")

# Detectar todos los archivos *_final.csv
files = [f for f in os.listdir(DATA_FINAL) if f.endswith("_final.csv")]
print(f"Archivos detectados: {files}")

# --- Función utilitaria para analizar cada tabla ---
def quick_eda(path):
    name = os.path.basename(path).replace("_final.csv", "")
    print(f"\n=== Análisis de {name.upper()} ===")
    df = pd.read_csv(path)
    print(f"Shape: {df.shape[0]} filas, {df.shape[1]} columnas\n")
    
    # Info básica
    display(df.head(3))
    
    # Valores nulos
    nulls = df.isna().sum()
    nulls_pct = (nulls / len(df) * 100).round(1)
    print("Valores nulos (%):")
    display(pd.DataFrame({"nulos": nulls, "%": nulls_pct}).sort_values("%", ascending=False).head(10))
    
    # Tipos de datos
    print("Tipos de datos:")
    display(df.dtypes.value_counts())

    # --- Distribución numérica general ---
    numeric_cols = df.select_dtypes(include=["number"]).columns
    if len(numeric_cols) > 0:
        print(f"\nColumnas numéricas ({len(numeric_cols)}): {list(numeric_cols[:10])}")
        desc = df[numeric_cols].describe().T
        desc["missing_%"] = df[numeric_cols].isna().mean() * 100
        display(desc.head(10))
        
        # Gráfico: histograma general (primeras 4 numéricas)
        sample_cols = numeric_cols[:4]
        df[sample_cols].hist(figsize=(10,6), bins=20)
        plt.suptitle(f"Distribución numérica ({name})", fontsize=14)
        plt.show()

    # --- Distribución categórica básica ---
    cat_cols = df.select_dtypes(include=["object"]).columns
    if len(cat_cols) > 0:
        for col in cat_cols[:2]:  # solo 2 para no sobrecargar
            top_vals = df[col].value_counts().head(10)
            plt.figure(figsize=(8,4))
            top_vals.plot(kind="bar", color="steelblue")
            plt.title(f"{name}: Top 10 valores en {col}")
            plt.xlabel(col)
            plt.ylabel("Frecuencia")
            plt.show()

    # --- Si tiene season o year, graficar temporalmente ---
    year_cols = [c for c in df.columns if "year" in c.lower() or "season" in c.lower()]
    for col in year_cols:
        plt.figure(figsize=(8,4))
        try:
            # Conversión segura a numérico
            numeric_series = pd.to_numeric(df[col], errors="coerce").dropna().astype(int)
            if not numeric_series.empty:
                numeric_series.value_counts().sort_index().plot(kind="bar", color="seagreen")
                plt.title(f"{name}: Conteo por {col}")
                plt.xlabel(col)
                plt.ylabel("Cantidad")
                plt.tight_layout()
                plt.show()
            else:
                print(f"{name}: la columna '{col}' no tiene valores numéricos válidos.")
        except Exception as e:
            print(f"Error procesando {col} en {name}: {e}")

    # --- Si tiene IDs duplicados ---
    id_cols = [c for c in df.columns if c.endswith("_id")]
    if id_cols:
        dup_count = df.duplicated(subset=id_cols).sum()
        if dup_count > 0:
            print(f"⚠️ Duplicados detectados en {id_cols}: {dup_count}")
        else:
            print(f"✅ Sin duplicados en claves {id_cols}")
    
    return df

# --- EDA sobre todas las tablas ---
datasets = {}
for file in files:
    path = os.path.join(DATA_FINAL, file)
    df = quick_eda(path)
    datasets[file] = df

print("\nEDA completo. Usa los gráficos y los conteos para evaluar la salud de cada tabla.")
```
